### Question 1. 
Convert all columns so that they use metric (International System of Units, SI) or derived units: temp and dewp to Celsius, precip to millimetres, visib to metres, as well as wind_speed and wind_gust to metres per second. Replace the data in-place (overwrite existing columns with new ones)

##### 1.1 Unit Conversion (Imperial to Metric)

To ensure consistency across our analyses, we’ll transform all pertinent weather measurements from imperial to metric (SI) or derived units. In this step, we will:

Convert temperature (temp) and dew point (dewp) from Fahrenheit to Celsius.

Change precipitation (precip) from inches to millimetres.

Translate visibility (visib) from miles to metres.

Recast wind speed (wind_speed) and wind gust (wind_gust) from miles per hour to metres per second.

These conversions establish a uniform unit framework that will underpin all downstream calculations and visualizations.


#### 1.2 The Code

```python
# Convert temperature and dew point from Fahrenheit to Celsius
for col in ['temp', 'dewp']:
    # (°F − 32) × 5/9 = °C
    df[col] = (df[col] - 32) * 5 / 9

# Convert precipitation from inches to millimetres (1 in = 25.4 mm)
df['precip'] = df['precip'] * 25.4

# Convert visibility from miles to metres (1 mi = 1609.34 m)
df['visib'] = df['visib'] * 1609.34

# Convert wind speed and gust from mph to m/s (1 mph = 0.44704 m/s)
for col in ['wind_speed', 'wind_gust']:
    df[col] = df[col] * 0.44704
```

### 3.1 Introduction:
### Daily Wind Speed Analysis

To examine LaGuardia Airport’s wind patterns throughout 2013, we’ll calculate and chart the daily average wind speed, which will spotlight both seasonal trends and unexpected spikes. Our workflow will:

*Select only the LGA observations from the full dataset

*Aggregate the data by calendar date to compute each day’s mean wind speed

*Populate any missing dates so that every day of 2013 appears in the series

*Render the results as a time-series line plot, using clear, well-formatted x-axis labels for ease of interpretation

### 3.2 The Code: 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def prepare_daily_mean(df: pd.DataFrame) -> pd.Series:
    """Filter to LGA, compute daily mean wind speed, and ensure all dates are present."""
    # Step 1: Filter for LaGuardia Airport
    df_lga = df[df['origin'] == 'LGA'].copy()

    # Step 2: Build a datetime column from year, month, day
    df_lga['date'] = pd.to_datetime(df_lga[['year', 'month', 'day']])

    # Step 3: Group by date and compute mean wind speed
    daily = df_lga.groupby('date')['wind_speed'].mean()

    # Step 4: Reindex to include every day of 2013
    all_days = pd.date_range(start='2013-01-01', end='2013-12-31', freq='D')
    daily = daily.reindex(all_days)
    daily.index.name = 'date'

    return daily

def plot_daily_mean(daily: pd.Series) -> None:
    """Plot the daily mean wind speed time series."""
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(daily.index, daily.values, linewidth=1)
    ax.set_title('Daily Mean Wind Speed at LGA (2013)')
    ax.set_xlabel('Date')
    ax.set_ylabel('Wind Speed (m/s)')

    # Format x-axis as one tick per month
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)

    ax.grid(True)
    plt.tight_layout()
    plt.show()

# Usage
daily_mean = prepare_daily_mean(df)
plot_daily_mean(daily_mean)
```

### 3.3 Conclusion

By isolating LGA’s records and calculating each day’s mean wind speed for 2013, we produced a concise time-series chart that brings both seasonal cycles and sudden wind spikes into focus. This visualization uncovers the underlying rhythm of wind behavior—highlighting routine patterns alongside days of exceptional calm or gustiness—and lays the groundwork for enhanced forecasting, anomaly detection, and strategic operational planning at the airport.


### Question 4. 
Identify the ten windiest days at LGA (dates and the corresponding mean daily wind speeds)

### 4.1 Introduction: 
#### Identifying the 10 Windiest Days at LGA

To deepen our understanding of LGA’s wind behavior, we’ll isolate the 10 days in 2013 with the highest average wind speeds, highlighting periods of extreme wind that could have affected safety or operations. The steps in the code below are to:

*Order the daily mean wind speeds from highest to lowest

*Extract the first ten entries in this sorted list

*Reset the index for a neat, sequential presentation of these top wind days


### 4.2 Code:

```python
def top_windiest_days(daily_series: pd.Series, n: int = 10) -> pd.DataFrame:
    """
    Return the top-n windiest days from a daily wind speed Series.
    
    Parameters:
    - daily_series: pd.Series indexed by date containing mean wind speeds.
    - n: Number of top entries to return.
    
    Returns:
    - DataFrame with columns ['date', 'wind_speed'], sorted by descending wind_speed.
    """
    # Reset index to turn the Series into a DataFrame
    df = daily_series.rename('wind_speed').reset_index()
    df.columns = ['date', 'wind_speed']
    
    # Sort descending and take the top n
    top_n = (
        df
        .sort_values(by='wind_speed', ascending=False)
        .head(n)
        .reset_index(drop=True)
    )
    return top_n

# Usage example
top_10_wind = top_windiest_days(daily_mean, n=10)
print(top_10_wind)
```

### 4.3 Conclusion

By ranking the daily averages from highest to lowest and pulling out the top ten, we’ve spotlighted the windiest days at LaGuardia Airport in 2013. This targeted list shines a light on the most extreme wind occurrences—data that can drive anomaly detection, help link wind extremes to delays or cancellations, and support detailed climatological research. The result is a concise table summarizing these peak wind events.

### Question 1. 
Compute the monthly mean wind speeds for all the three airports.
There is one obvious outlier amongst the observed wind speeds. Locate it (programmatically, do
not hardcode the date/day/row number) and replace it with np.nan (NaN) before computing the
means.

### 1.1 Introduction: 

In this section, we’ll derive each airport’s monthly average wind speed (for EWR, JFK, and LGA) with outlier protection by:

*Identifying any wind-speed readings that exceed four standard deviations from the global mean.

*Masking those extreme values as NaN so they’re omitted from our calculations.

*Grouping the cleaned data by airport and month, then computing the mean wind speed for each subgroup.

### 1.2 The code: 

```python
import numpy as np
import pandas as pd

# Step 1: Compute mean and standard deviation of wind_speed (ignoring existing NaNs)
valid_ws = df['wind_speed'].dropna()
mean_ws, std_ws = valid_ws.mean(), valid_ws.std()

# Step 2: Define 4-σ bounds to flag extreme outliers
lower, upper = mean_ws - 4 * std_ws, mean_ws + 4 * std_ws

# Step 3: Locate and report the outlier(s)
outlier_mask = (df['wind_speed'] < lower) | (df['wind_speed'] > upper)
outliers = df.loc[outlier_mask, ['origin', 'year', 'month', 'day', 'wind_speed']]
print("Identified outlier(s):")
print(outliers)

# Step 4: Replace outliers with NaN (so they’re ignored in aggregation)
df.loc[outlier_mask, 'wind_speed'] = np.nan

# Step 5: Compute monthly mean wind speeds for each airport (NaNs excluded)
monthly_mean = (
    df
    .groupby(['origin', 'month'], as_index=False)['wind_speed']
    .mean()
    .rename(columns={
        'origin': 'Airport',
        'month': 'Month',
        'wind_speed': 'Monthly Mean Wind Speed (m/s)'
    })
)

# Step 6: Display the first 12 rows of the result
print("\nMonthly mean wind speeds:")
print(monthly_mean.head(12))
```

#### 1.3 Conclusion
By marking the one extreme wind-speed outlier (more than four standard deviations away) as NaN, we prevented it from skewing our results. The cleaned monthly_mean table now accurately captures the true month-to-month variations in average wind speed for each airport over 2013.

### Question 2:
Draw the monthly mean wind speeds for the three airports on the same plot (three curves of dif-

### 2.1 Introduction

To compare wind seasonality at LGA, JFK, and EWR in 2013, we first cleansed the data by treating any hourly wind‐speed reading more than four standard deviations from the mean as missing. After excluding these outliers, we calculated each airport’s average wind speed by calendar month. We then:

*Constructed a datetime index (one entry per month) for plotting.

*Pivoted the table so that each airport formed its own series.

*Plotted all three series together on a single line chart, complete with clear labels and formatting, to visualize and contrast their year-round wind patterns.


```python
def plot_monthly_wind_comparison(monthly_mean: pd.DataFrame) -> None:
    """
    Pivot the monthly_mean DataFrame so each airport is a column,
    then plot their monthly mean wind speeds.
    
    Parameters:
    - monthly_mean: DataFrame with columns ['Airport', 'Month', 'Monthly Mean Wind Speed (m/s)']
    """
    # Pivot so each airport is its own series
    pivot_df = monthly_mean.pivot(
        index='Month',
        columns='Airport',
        values='Monthly Mean Wind Speed (m/s)'
    )

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(8, 4))

    # Plot each airport’s monthly series
    for airport in pivot_df.columns:
        ax.plot(pivot_df.index, pivot_df[airport], marker='o', label=airport)

    # Set labels, title, and legend
    ax.set_xlabel('Month')
    ax.set_ylabel('Monthly Mean Wind Speed (m/s)')
    ax.set_title('Monthly Mean Wind Speeds by Airport (2013)')
    ax.legend(title='Airport', loc='best')

    # Ensure months 1–12 are shown on the x-axis
    ax.set_xticks(pivot_df.index)
    ax.set_xticklabels(pivot_df.index)

    # Improve layout and display
    plt.tight_layout()
    plt.show()

# Usage
plot_monthly_wind_comparison(monthly_mean)
```

### 2.3 Conclusion:

The chart highlights pronounced seasonal rhythms: all three airports see wind speeds crest in late winter and spring, then taper off during the summer. Although they share a similar annual cycle, subtle distinctions appear—JFK consistently records the strongest winds, while EWR remains the calmest. These observations can guide operational decisions (for example, timing maintenance for the calmer months) and motivate further study into the meteorological drivers behind each airport’s unique wind profile.
